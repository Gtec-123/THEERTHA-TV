```python
import cv2
```

```python
import os
```

```python
data_path='C:\\Users\\software 2\\Desktop\\Theertha DL\\Fruits_Vegetables\\Fruits_Vegetables\\train'
```

```python
categories=os.listdir(data_path) #returns a list of filenames in the specified directory
```

```python
labels=[i for i in range(len(categories))]
label_dict=dict(zip(categories,labels))
```

```python
print(label_dict)
```

```python
print(categories)
```

```python
print(labels)
```

```python
img_size=100
```

```python
data=[]
```

```python
target=[]
```

```python
for category in categories:
 folder_path=os.path.join(data_path,category) #joins the specified paths data_path and category
 img_names=os.listdir(folder_path)
 for img_name in img_names:
  
  img_path=os.path.join(folder_path,img_name)
  img=cv2.imread(img_path) # to read an image from a file specified Â path
  try:
   gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #convert image to gray scale
   resized=cv2.resize(gray,(img_size,img_size)) #resizing the grayscale into 100*100,#since we need a fixed common size for all the images
   data.append(resized)
   target.append(label_dict[category]) #appending the image and the label (categorized) into the list (dataset)
  except Exception as e:
    print('Exception:',e)  #if there is any exception will be printed here and pass to the next image
 

 
 

```

```python
import numpy as np
data=np.array(data)
```

```python
print(data.shape) 
```

```python
data=np.reshape(data,(data.shape[0],img_size,img_size,1)) 
```

```python
target=np.array(target)
```

```python
from keras.utils import to_categorical

```

```python
new_target=to_categorical(target) #to convert class vectors (integers) to binary class matrices,new target will be a 2D array where each row corresponds to a sample, and each column corresponds to a class. The column for the true class will be set to 1, and all other columns will be set to 0.
```
